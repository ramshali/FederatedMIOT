```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("SpirometerObservations_v1.3.csv")

# Display the first few rows of the dataset
print(data.head())

# 1. Difference Between Observed and Reference Values
# Separate data for males and females
males = data[data['Gender'] == 'male']
females = data[data['Gender'] == 'female']

# Plot for FEV1
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FEV1 (litres)', data=data, color='blue', label='Observed FEV1')
sns.barplot(x='Gender', y='FEV1 Ref Value', data=data, color='orange', label='Reference FEV1', alpha=0.5)
plt.title('FEV1: Observed vs Reference Values')
plt.ylabel('FEV1 (litres)')
plt.legend()
plt.show()

# Plot for FVC
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FVC (litres)', data=data, color='green', label='Observed FVC')
sns.barplot(x='Gender', y='FVC Ref Value', data=data, color='red', label='Reference FVC', alpha=0.5)
plt.title('FVC: Observed vs Reference Values')
plt.ylabel('FVC (litres)')
plt.legend()
plt.show()

# Plot for FER%
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FER (%)', data=data, color='purple', label='Observed FER%')
sns.barplot(x='Gender', y='FER Ref. Value', data=data, color='yellow', label='Reference FER%', alpha=0.5)
plt.title('FER%: Observed vs Reference Values')
plt.ylabel('FER (%)')
plt.legend()
plt.show()

# Plot for PEF
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='PEF (litre/min)', data=data, color='cyan', label='Observed PEF')
sns.barplot(x='Gender', y='PEF Ref. Value', data=data, color='magenta', label='Reference PEF', alpha=0.5)
plt.title('PEF: Observed vs Reference Values')
plt.ylabel('PEF (litre/min)')
plt.legend()
plt.show()

# 2. Impact of PM2.5 on Lung Function
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.scatterplot(x='PM2.5', y='FEV1 (litres)', data=data, hue='Gender')
plt.title('PM2.5 vs FEV1')

plt.subplot(2, 2, 2)
sns.scatterplot(x='PM2.5', y='FVC (litres)', data=data, hue='Gender')
plt.title('PM2.5 vs FVC')

plt.subplot(2, 2, 3)
sns.scatterplot(x='PM2.5', y='FER (%)', data=data, hue='Gender')
plt.title('PM2.5 vs FER%')

plt.subplot(2, 2, 4)
sns.scatterplot(x='PM2.5', y='PEF (litre/min)', data=data, hue='Gender')
plt.title('PM2.5 vs PEF')

plt.tight_layout()
plt.show()

# 3. Impact of Daytime on Lung Function
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='morning/afternoon/evening', y='FEV1 (litres)', data=data, hue='Gender')
plt.title('FEV1 by Time of Day')

plt.subplot(2, 2, 2)
sns.boxplot(x='morning/afternoon/evening', y='FVC (litres)', data=data, hue='Gender')
plt.title('FVC by Time of Day')

plt.subplot(2, 2, 3)
sns.boxplot(x='morning/afternoon/evening', y='FER (%)', data=data, hue='Gender')
plt.title('FER% by Time of Day')

plt.subplot(2, 2, 4)
sns.boxplot(x='morning/afternoon/evening', y='PEF (litre/min)', data=data, hue='Gender')
plt.title('PEF by Time of Day')

plt.tight_layout()
plt.show()

# 4. Accuracy of Devices (SP80B and MicroPlus)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='Device', y='FEV1 (litres)', data=data)
plt.title('FEV1 by Device')

plt.subplot(2, 2, 2)
sns.boxplot(x='Device', y='FVC (litres)', data=data)
plt.title('FVC by Device')

plt.subplot(2, 2, 3)
sns.boxplot(x='Device', y='FER (%)', data=data)
plt.title('FER% by Device')

plt.subplot(2, 2, 4)
sns.boxplot(x='Device', y='PEF (litre/min)', data=data)
plt.title('PEF by Device')

plt.tight_layout()
plt.show()
```
