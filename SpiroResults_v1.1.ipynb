```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("SpirometerObservations_v1.3.csv")

# Display the first few rows of the dataset
print(data.head())

# Function to save plots
def save_plot(plot, filename):
    plot.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved {filename}")

# Set a consistent color palette
#observed_color = '#1f77b4'  # Blue for observed values
#reference_color = '#ff7f0e'  # Orange for reference values

# 1. Difference Between Observed and Reference Values
# Plot for FEV1
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FEV1 (litres)', data=data, color='blue', label='Observed FEV1')
sns.barplot(x='Gender', y='FEV1 Ref Value', data=data, color='none', edgecolor="orange", hatch="//", label='Reference FEV1', alpha=0.6)
plt.title('FEV1: Observed vs Reference Values')
plt.ylabel('FEV1 (litres)')
plt.legend()
save_plot(plt, 'FEV1_Observed_vs_Reference.png')
plt.show()

# Plot for FVC
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FVC (litres)', data=data, color='green', label='Observed FVC')
sns.barplot(x='Gender', y='FVC Ref Value', data=data, color='none', edgecolor="orange", hatch="//", label='Reference FVC', alpha=0.6)
plt.title('FVC: Observed vs Reference Values')
plt.ylabel('FVC (litres)')
plt.legend()
save_plot(plt, 'FVC_Observed_vs_Reference.png')
plt.show()

# Plot for FER%
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='FER (%)', data=data, color='purple', label='Observed FER%')
sns.barplot(x='Gender', y='FER Ref. Value', data=data, color='none', edgecolor="yellow", hatch="//", label='Reference FER%', alpha=0.6)
plt.title('FER%: Observed vs Reference Values')
plt.ylabel('FER (%)')
plt.legend()
save_plot(plt, 'FER_Observed_vs_Reference.png')
plt.show()

# Plot for PEF
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='PEF (litre/min)', data=data, color='cyan', label='Observed PEF')
sns.barplot(x='Gender', y='PEF Ref. Value', data=data, color='none', edgecolor="magenta", hatch="//", label='Reference PEF', alpha=0.6)
plt.title('PEF: Observed vs Reference Values')
plt.ylabel('PEF (litre/min)')
plt.legend()
save_plot(plt, 'PEF_Observed_vs_Reference.png')
plt.show()
```
